Import any necessary libraries to perform Isomap here, reduce `df` down to three components and using `K=6` for your neighborhood size:

```python
from sklearn import manifold
iso = manifold.Isomap(n_neighbors=10, n_components=3)
iso.fit(df)
manifold.Isomap(eigen_solver='auto', max_iter=None, n_components=3, n_neighbors=10,neighbors_algorithm='auto', path_method='auto', tol=0)
manifold = iso.transform(df)
print(df.shape)
manifold.shape
```

```python
manifold_df=pd.DataFrame(manifold)
manifold_df.columns=['First','Second','Third']
manifold_df.head()
```
